### **7.1 Plotting the Distribution of Hogwarts Houses**

Our first spell is to visualize the distribution of students across the four houses of Hogwarts: Gryffindor, Hufflepuff, Ravenclaw, and Slytherin. This chart is akin to the Sorting Hat ceremony, where each student's house is revealed, a moment filled with anticipation and excitement.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Path to our dataset
dataset_path = 'data/hogwarts-students-02.csv'

# Reading the dataset
hogwarts_df = pd.read_csv(dataset_path)

# Plotting the distribution of Hogwarts Houses with count labels
plt.figure(figsize=(10, 6))
sns.countplot(x='house', hue='house', data=hogwarts_df, palette='viridis')
plt.title('Distribution of Students Across Hogwarts Houses')
plt.xlabel('House')
plt.ylabel('Number of Students')

# Add count labels on top of bars (using Matplotlib)
for container in plt.gca().containers:
    plt.bar_label(container)

# Explicitly create the legend
plt.legend(title='House')

plt.show()
```

_Gaze_ upon this chart! It reveals the composition of each house, showing whether Gryffindor is brimming with brave souls or if Ravenclaw has an influx of wise minds. This visualization helps us understand the balance of traits across the school, much like observing the diverse personalities that roam the halls of Hogwarts.

### **7.2 Visualizing Relationships between Features**

Next, we weave a more intricate spell, exploring the relationships between different features in our dataset. For instance, does a student‚Äôs heritage influence their choice of pet, or is there a connection between a student‚Äôs age and the type of wand they use? This step is like exploring the Forbidden Forest, uncovering the connections and mysteries that lie within.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Path to your dataset
dataset_path = 'data/hogwarts-students-02.csv'

# Reading the dataset
hogwarts_df = pd.read_csv(dataset_path)

# Plotting the distribution of Hogwarts Houses with student counts
plt.figure(figsize=(10, 5))
sns.countplot(x='house', hue='pet', data=hogwarts_df, palette='viridis')

# Add data labels (student counts) on top of each bar
for container in plt.gca().containers:
    plt.bar_label(container)

plt.title('Relationship between "House" and "Choice of Pet"')
plt.xlabel('House')
plt.ylabel('Number of Students')
plt.legend(title='Pet Type')
plt.show()
```

Through this visualization, we might discover that Muggle-born students have a penchant for owls, while Pure-bloods prefer cats. These insights are akin to understanding the habits of magical creatures, revealing the subtle nuances that define the Hogwarts community.

### **7.3 Interactive Charts for an Engaging Experience**

For a truly magical experience, we delve into the realm of interactive charts. These dynamic visualizations allow us to engage with the data in real-time, much like flipping through the pages of a spellbook that writes itself. Tools like Plotly and Bokeh provide this interactivity, turning our static charts into living, breathing entities.

```python
try:
  import plotly.express as px
except ModuleNotFoundError:
  !pip install plotly.express  # Install Plotly if not already available

# Assuming successful installation or existing Plotly library
import pandas as pd
import plotly.express as px

# Path to your dataset
dataset_path = 'data/hogwarts-students-02.csv'

# Reading the dataset
hogwarts_df = pd.read_csv(dataset_path)

# Creating an interactive chart for Age distribution across Houses
fig = px.histogram(hogwarts_df, x='age', color='house', barmode='overlay',
                  title='Age Distribution Across Hogwarts Houses',
                  labels={'Age': 'Student Age', 'House': 'Hogwarts House'})

fig.show()
```

_Behold_! This chart allows us to hover over different bars, revealing detailed information about the age distribution within each house. It‚Äôs like using the Marauder‚Äôs Map to track the movements of every student within the castle, offering a deeper level of engagement and understanding.

### **7.5 Discovering Patterns with Pair Plots** 

To uncover the deeper mysteries of our data, we turn to pair plots. These powerful visualizations allow us to see how multiple features interact with one another, much like observing the interplay of spells in a duel.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Path to your dataset
dataset_path = 'data/hogwarts-students-02.csv'

# Reading the dataset
hogwarts_df = pd.read_csv(dataset_path)

# Setting a larger figure size for the pairplot
plt.figure(figsize=(30, 15))  # Adjust width and height as needed

# Creating pair plots with color-coded houses
sns.pairplot(hogwarts_df, hue='house', palette='husl')
plt.title('')
plt.show()
```

In this visualization, each plot reveals a different aspect of the dataset, from the relationship between age and wand type to the correlation between heritage and house. It's like seeing through the eyes of a Seer, gaining insight into the interconnectedness of the magical world.
